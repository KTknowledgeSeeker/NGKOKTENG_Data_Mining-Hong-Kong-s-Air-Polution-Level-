The algorithm implements by recursively choosing the best features from the dataset to separate the data (step 7: root.test_cond = find_best_split (E, F)) and extending the leaf of the decision tree ((step 11: child TreeeGrowth (E, F).) and (step 12: let V = {V|V is a possible outcome of root.test_cond})). The details and pseudo-code of the decision tree algorithm are briefed below:

1. The function of the step of createNode() expands the size of the decision tree classifier by establishing a new node (branch) in the decision tree model. A node in the decision tree classifier either a test condition, represented as the code of node.test_cond, or a class label, represented as the code of node.label.

2. The function of the find_best_split() is used to ensure which feature should be chosen as the test condition for splitting the training records from the dataset. There has mentioned in the previous section, the concentration of the test condition based on which non-purity measured is used to ensure the best of a split. Some widely used measures include entropy, the Gini index, and the x^2 statistic.

3. For the Classify() function identifies which class label to be distributed to the leaf node part. Each of these leaf node t, p(i|t) will refers to the fraction of training results from class i related with the node t. Recently, leaf node is setup for the class which has lots of training records: 
            leaf.label = argmax p(i|t)
   while the argmax operator will then return the i argument which will maximizes the p(i|t) expression. On the other hand, it will provide the information for identifying the class label of the leaf node. Also, p(i|t) is able to predict the prediction for the records which is for leaf node t to refers to the class i. 

4. For the stopping_cond(), it is used to stop the tree-growing continuously. It is processed by checking if all the records are having thr similar class label or having the similar feature values. Otherthan that, it can be stopped by checking if the records have been lower than the minimum threshold. 

After building the whole decision tree, tree-pruning will be triggered to minimize the sizes of the decision tree. Overfitting will be assigned to the decision trees which are having too large of records. However, the tree-pruning will help to trim the branches generated from the decision tree while improving the decision tree's generalization.

###### Introduction of the Dataset

The dataset that had been used in our project was collected from the World Air Quality Index project (AQICN), there has shown a restricted dataset composes of daily recording of the air pollution traffic flows (at local data collection and air detection stations). The gas and materials that have been considered will cause the air pollution are pm2.5, pm10, so2, no2, and o3, within the Hong Kong local urban area for the year from 2014 January to 2021 February 2nd. The air pollution traffic flow data were collected over the 9 air quality detection station locations with additional five parameters including pm2.5, pm10, so2, no2, and o3. Besides, there had also added some extra features in the dataset (csv file), such as date, day, year, month, city, country, city_shortform, and country_shortform, which is easy to let the user know when is the record and where is the location.
 
Parameter units that have been used for measuring the elements of pm2.5 and pm10 are micrograms/cubic meter, the parameter unit that had been used for measuring the elements of so2, no2, and o3 are parts-per-billion-volume (ppbV).
 
After the step of data integration, there had merged the dataset from nine air quality detection station into the dataset. Besides, there has also dropped the feature of date because there has already shown the features of the day, month, and year, so the user can only remo"ve those feature to know when was the air quality record. In addition, there have added two new columns into the final dataset, which is called index (Air Quality Index) and quality (Air Quality Level). The way to measure the index is there will take the daily highest record from the elements of pm2.5, pm10, so2, no2, and o3, then the highest value will become the index of the day. The quality (Air Quality Level) has six different levels such as ‘good’, ‘moderate’, ‘unhealthy for sensitive group’, ‘unhealthy’, ‘very unhealthy’, and ‘hazardous’, the quality of ‘good’ is between the index of 0 and 50, the quality of ‘moderate’ is between the index of 50 and 100, and so on.

```python
# Create a single column 
quality = pd.Series([]) 

x=0;

# Determine the air quality index level

for x in range(len(dataFrame)) :
    if ((dataFrame["index"][x] >= 0) and (dataFrame["index"][x] <= 50)):
        quality[x]="Good"
    elif ((dataFrame["index"][x] > 50) and (dataFrame["index"][x] <= 100)) :
        quality[x]="Moderate"
    elif ((dataFrame["index"][x] > 100) and (dataFrame["index"][x] <= 150)) :
        quality[x]="Unhealthy for Sensitive Group"
    elif ((dataFrame["index"][x] > 150) and (dataFrame["index"][x] <= 200)) :
        quality[x]="Unhealthy"
    elif ((dataFrame["index"][x] > 200) and (dataFrame["index"][x] <= 300)) :
        quality[x]="Very Unhealthy"
    elif ((dataFrame["index"][x] > 300) and (dataFrame["index"][x] <= 500)) :
        quality[x]="Hazardous"

# Insert the single column into the dataFrame
dataFrame.insert(13, "quality", quality) 
```

```python
# Split dataset in features and target variable
feature_cols = [' pm25', ' pm10', ' o3', ' no2',' so2']
# Features
X = dataFrame[feature_cols] 
# Target variable
y = dataFrame.quality
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Print the classification report of decision tree
print(classification_report(y_test, y_pred))
```

```python
# Get the accurancy score by using Decision Tree to perform the prediction
score_dtc = accuracy_score(y_pred,y_test)*100
print("The accuracy score by using Decision Tree to perform the prediction is: ", score_dtc," %")
```
